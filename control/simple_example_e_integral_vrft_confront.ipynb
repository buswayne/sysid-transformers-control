```python
ts = 1e-2
T = 20#seq_len*ts
t_CL = np.arange(0, T, ts)
r_batch = np.zeros((batch_size, len(t_CL)))
# Generate the reference
generate_from_model = False
if generate_from_model:
    test_ds = SimpleExample1Dataset(seq_len=len(t_CL), normalize=True, return_y=True)
    test_dl = DataLoader(test_ds, batch_size=batch_size)
    batch_u, batch_e, batch_y, batch_r = next(iter(test_dl))
    for i in range(r_batch.shape[0]):
        # get virtual error
        r_batch[i,:] = batch_r[i,:,0]
else:
    r = np.zeros(t_CL.shape)
    r[t_CL < 10] = 10
    r[(t_CL >= 2.5) & (t_CL < 5)] = -10
    r[(t_CL >= 5.0) & (t_CL < 7.5)] = 0
    r[(t_CL >= 7.5) & (t_CL < 10)] = 5
    r[(t_CL >= 10.) & (t_CL < 12.5)] = -5
    r[(t_CL >= 12.5) & (t_CL < 15.)] = 20
    r[(t_CL >= 15.0) & (t_CL < 17.5)] = 0
    r[(t_CL >= 17.5) & (t_CL < 20.0)] = -20
    for i in range(r_batch.shape[0]):
        # get virtual error
        r_batch[i,:] = r
```

```python
s = tf('s')
z = tf([1, 0], [1], dt=ts)
tau = 0.5# s
M = 1/(1 + (tau/(2*np.pi))*s)
M = M * (1 + 1e-2 * (tau / (2 * np.pi)) * s)
#M = c2d(M, ts, 'matched')

y_d = lsim(M,r,t_CL)[0]

plt.plot(r_batch[0,:])
plt.plot(y_d)
plt.legend(['$ref$', '$y_d$'])
```

```python
n_testing = 3
H = 5
Mr_E = np.zeros((n_testing, len(t_CL)))

y_log = np.zeros((n_testing, len(t_CL)))
e_log = np.zeros((n_testing, len(t_CL)))
e_int_log = np.zeros((n_testing, len(t_CL)))
u_log = np.zeros((n_testing, len(t_CL)))

Ts = 1e-2
Fs = 1 / Ts
s = tf('s')
# problem data

with torch.no_grad():
    for test_idx in range(n_testing):  # fixed the variable name to avoid confusion
        print(test_idx)
        data = data_test[test_idx]

        # system
        num = [data['num_1']]
        den = [data['den_1'], data['den_2'], data['den_3']]
        P = tf(num, den)

        # Initialize u, e, e_int with the correct shape to avoid concatenation issues
        u = torch.zeros((1, 0, 1))
        e = torch.zeros((1, 0, 1))
        e_int = torch.zeros((1, 0, 1))

        # This is like t=0, for our model we need u(-1).
        u_i = 1e-6  # u(-1)
        x_i = np.array([[1e-6, 1e-6]])
        y_i = 1e-6
        e_int_i = 1e-6

        u_log[test_idx, 0] = u_i
        y_log[test_idx, 0] = y_i

        for i in range(0, len(t_CL)):
            u_i_tensor = torch.Tensor([[[u_i]]])  # first you add u(t)
            #print(f"u_i_tensor shape: {u_i_tensor.shape}, u shape: {u.shape}")
            u = torch.cat((u, u_i_tensor / 1000), 1)  # concatenate along the correct dimension
            u = u.reshape(1, -1, 1)

            e_i_tensor = torch.Tensor([[[r_batch[test_idx, i] - y_i]]])  # e(t)
            #print(f"e_i_tensor shape: {e_i_tensor.shape}, e shape: {e.shape}")
            e = torch.cat((e, e_i_tensor / 0.3), 1)  # concatenate along the correct dimension
            e = e.reshape(1, -1, 1)

            e_int_i_tensor = torch.Tensor([[[r_batch[test_idx, i] - y_i + e_int_i]]])
            #print(f"e_int_i_tensor shape: {e_int_i_tensor.shape}, e_int shape: {e_int.shape}")
            e_int = torch.cat((e_int, e_int_i_tensor / 220), 1)  # concatenate along the correct dimension
            e_int = e_int.reshape(1, -1, 1)

            input = torch.cat((e, e_int), 2)

            if i < H:
                pred, _ = model(input.reshape(1, -1, 2), u.reshape(1, -1, 1))
            else:
                input_tmp = input.reshape(1, -1, 2)
                u_tmp = u.reshape(1, -1, 1)
                pred, _ = model(input_tmp[:, -H:, :], u_tmp[:, -H:, :])

            u_i = pred[:, -1, :].item() * 1000  # u(t)

            # Controller u(t) = C(e(t-1), u(t-1))

            y_i, _, x_i = lsim(P, u_i, np.array([0.0, Ts]), x_i[-1], return_x=True)
            y_i = y_i[-1]  # y(t) <- y(t+1)

            e_log[test_idx, i] = e_i_tensor.item()
            e_int_log[test_idx, i] = e_int_i_tensor.item()

            if (i + 1) < len(t_CL):
                u_log[test_idx, i + 1] = u_i
                y_log[test_idx, i + 1] = y_i

        Mr_E[test_idx] = y_d - y_log[test_idx]
```

```python
fig = plt.figure(figsize=(7, 5))
#plt.plot(t_CL, r)
plt.plot(t_CL, r_batch[0,:], label="$r$", c='tab:orange')
plt.plot(t_CL, y_log[0,:].T, label="$y$", c='tab:blue', alpha=1)
plt.plot(t_CL, y_d, label="$y_d}$", c='tab:green')
plt.legend(['$r$','$y$', '$y_{d}$'])
plt.tick_params('x', labelbottom=False)
```

```python
## CALCULATE MSE (r, y) 
error_tot = 0
for i in range(n_testing):
     for k in range(len(e_log[i])):
        squared_error = e_log[i,k] * e_log[i,k]
        error_tot += squared_error
         
MSE = error_tot / (len(e_log[0]) * (n_testing+1))
print(f"MSE: {MSE:.5f}")

Mr_error_tot = 0
for i in range(n_testing):
     for k in range(len(Mr_E[i])):
        mr_squared_error = Mr_E[i,k] * Mr_E[i,k]
        Mr_error_tot += mr_squared_error
         
MSE2 = Mr_error_tot / (len(Mr_E[0]) * (n_testing+1))
print(f"MSE  (y_d,y): {MSE2:.5f}")



# relative error, in percentage
#relative_error = 0
#counter_discarded_point = 0
#for i in range(n_testing+1) : 
#    for k in range(len(e_log[i])):
#        e_abs = np.abs(e_log[i,k])
#        ref_abs = (np.abs(r_batch[i,k]))
#        e_rel = e_abs / ref_abs
#        if e_rel < 20 : # samples that are too big are caused by divisions by zero, unwanted
#            relative_error += e_rel
#        else : counter_discarded_point += 1
#re = relative_error/ (( len(e_log[0]) * (n_testing+1))-counter_discarded_point) * 100
#print(f"Relative Error: {re:.2f}%")
```
